```python
# Importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Importing dataset
df = pd.read_csv('gt.csv', encoding="ISO-8859-1", low_memory = False)
df.head()
```

```python
df.shape #This means Dataset has 181691 Datapoints and 135 Columns i.e, Categories
```

```python
df.describe() # Description
```

```python
df.columns.values #All column names
```

```python
df.rename(columns={'iyear':'Year','imonth':'Month','iday':'Day','country_txt':'Country','provstate':
                  'State','region_txt':'Region','attacktype1_txt':'AttackType','target1':'Target',
                  'nkill':'Killed','nwound':'Wounded','summary':'Summary','gname':'Group',
                   'targtype1_txt':'Target_Type','weaptype1_txt':'Weapon_type','motive':'Motive',
                  'extended':'Extended','latitude':'Latitude','longitude':'Longitude', 'target1': 'Target','city':'City'}, inplace = True)
```

```python
df.columns.values #Changes the name of few columns
```

```python
#Now will check how many null values in dataset

df.isnull().sum()
```

```python
#Total number of null values

df.isnull().sum().sum()
```

```python
# Taking only necessary columns in data frame
df = df[['Year','Month','Day','Country','State','Region','City','Latitude','Longitude',
                 'AttackType','Killed','Wounded','Group','Target_Type','Weapon_type']]
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Killed']
```

```python
df['Wounded']
```

```python
#convert NaN values in Killed and Wounded to 0

df['Wounded'] = df['Wounded'].fillna(0).astype(int)
df['Killed'] = df['Killed'].fillna(0).astype(int)
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df[df.State.isnull()]
```

```python
df=df.dropna(subset=['State'])  #Dropping all Datapoints with NaN values
```

```python
df=df.dropna(subset=['City'])
```

```python
df=df.dropna(subset=['Longitude'])
```

```python
df=df.dropna(subset=['Latitude'])
```

```python
df.isnull().sum() #Now dataset has no NaN values
```

```python
df.shape #New Shape of the Dataset
```

```python
df.corr()
```

```python
# Heatmap to check collinearity
import seaborn as sns

plt.figure(figsize=(15,7))
sns.heatmap(df.corr(), annot= True)
plt.show()
```

```python
df.nunique()
```

```python
# Regions with most attacks
plt.figure(figsize = (20,10))
sns.barplot(x=df['Region'].value_counts().index, y=df['Region'].value_counts().values)
plt.title('Most Attacked Regions')
plt.xlabel('Regions')
plt.ylabel('Number of Attacks')
plt.xticks(rotation= 90)
plt.show()
```

```python
# countries with most attacks

plt.figure(figsize = (20,10))
sns.barplot(x=df['Country'].value_counts()[:10].index, y=df['Country'].value_counts()[:10].values)
plt.title('Most Attacked Countries')
plt.xlabel('Countries')
plt.ylabel('Number of Attacks')
plt.xticks(rotation= 90)
plt.show()
```

```python
#State with most attacks

plt.figure(figsize = (20,10))
sns.barplot(x=df['State'].value_counts()[:10].index, y=df['State'].value_counts()[:10].values,palette="spring")
plt.title('Most Attacked States')
plt.xlabel('States')
plt.ylabel('Number of Attacks')
plt.xticks(rotation= 90)
plt.show()
```

```python
#Most attacked City
plt.figure(figsize = (20,10))
sns.barplot(x=df['City'].value_counts()[:10].index, y=df['City'].value_counts()[:10].values,palette="summer")
plt.title('Most Attacked Cities')
plt.xlabel('Cities')
plt.ylabel('Number of Attacks')
plt.xticks(rotation= 90)
plt.show()
```

```python
#Month vs Attack

plt.figure(figsize = (20,10))
sns.barplot(x=df['Month'].value_counts().index, y=df['Month'].value_counts().values,palette="winter")
plt.title('Most Attacked Months')
plt.xlabel('Months')
plt.ylabel('Number of Attacks')
plt.xticks(rotation= 90)
plt.show()
```

```python
# No. of terrorist attacks wrt year
plt.figure(figsize = (20,10))
sns.barplot(x=df['Year'].value_counts().index, y=df['Year'].value_counts().values)
plt.title('Terror Attack over the years')
plt.xlabel('Years')
plt.ylabel('Number of Attacks')
plt.xticks(rotation= 90)
plt.grid()
plt.show()
```

```python
#Attack Type Analysis

plt.figure(figsize = (20,10))
sns.barplot(x=df['AttackType'].value_counts().index, y=df['AttackType'].value_counts().values,palette="autumn")
plt.title('Attack Type Analysis')
plt.xlabel('Attack Type')
plt.ylabel('Number of Attacks')
plt.xticks(rotation= 90)
plt.show()
```

```python
#Target Type Analysis
plt.figure(figsize = (20,10))
sns.barplot(x=df['Target_Type'].value_counts().index, y=df['Target_Type'].value_counts().values,palette='rocket_r')
plt.title('Target Type Analysis')
plt.xlabel('TargetType')
plt.ylabel('Number of Attacks')
plt.xticks(rotation= 90)
plt.show()
```

```python
#Weapon Type Analysis

plt.figure(figsize = (20,10))
sns.barplot(x=df['Weapon_type'].value_counts().index, y=df['Weapon_type'].value_counts().values,palette='viridis')
plt.title('Weapon Type Analysis')
plt.xlabel('WeaponType')
plt.ylabel('Number of Attacks')
plt.xticks(rotation= 90)
plt.show()
```

```python
# Killed vs Region
plt.figure(figsize = (15,7))
df.groupby(['Region'])['Killed'].sum().sort_values(ascending = False).head(10).plot(kind = 'bar', colormap = 'seismic')
plt.xticks(rotation=90)
plt.title('No. of people killed Region wise')
plt.ylabel("Killed")
plt.show()
```

```python
# Wounded vs Region
plt.figure(figsize = (15,7))
df.groupby(['Region'])['Wounded'].sum().sort_values(ascending = False).head(10).plot(kind = 'bar', colormap = 'seismic')
plt.xticks(rotation=90)
plt.title('No. of people wounded Region wise')
plt.ylabel("Wounded")
plt.show()
```

```python
# Killed vs Country
plt.figure(figsize = (15,7))
df.groupby(['Country'])['Killed'].sum().sort_values(ascending = False).head(10).plot(kind = 'bar', colormap = 'PRGn')
plt.xticks(rotation=90)
plt.title('No. of people killed Country wise')
plt.ylabel("Killed")
plt.show()
```

```python
# Wounded vs Country
plt.figure(figsize = (15,7))
df.groupby(['Country'])['Wounded'].sum().sort_values(ascending = False).head(15).plot(kind = 'bar', colormap = 'PRGn')
plt.xticks(rotation=90)
plt.title('No. of people wounded Country wise')
plt.ylabel("Wounded")
plt.show()
```

```python

#Terror Groups
plt.figure(figsize = (15,7))
sns.barplot(x = df['Group'].value_counts()[1:11].index, y =df['Group'].value_counts()[1:11].values,palette='magma')
plt.title('Top 10 Terrorist Organisation with Highest Terror Attacks')
plt.ylabel('Terror Group Name')
plt.xlabel('Number of Attacks')
plt.xticks(rotation= 90)
plt.show()
```
